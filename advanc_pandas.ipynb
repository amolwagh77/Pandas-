#### merge operation  using pd.merge()

```python
data1=pd.DataFrame({'key1':[1,2,3,4,5],
       'key2':[6,7,8,5,6],
       'key3':[55,88,99,66,33]})
data1
```

```python
data2=pd.DataFrame({'key1':[1,2,3,4,5],
       'key4':[6,78,89,5,64],
       'key5':[55,88,9,6,3]})
data2
```

```python
pd.merge(data1,data2)
```

```python
pd.merge(data1,data2,how='left')
```
